```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('CaseAge.csv')
sns.set_theme(context='notebook', style='whitegrid')
```

# Boxplots for each tech queue, per severity level
This now automated process was very time consuming and tedious in order to create new box plots each month as defect cases are closed and a tech case SLA became available. Previously, each boxplot was created manually then inserted into a powerpoint-- manually selecting columns and genenerating 1 chart and formatting it to insert into the powerpoint took anywhere from 5-8 minutes, depending on volume of data and column location within the csv.

```python
sns.boxplot(data=df.filter(regex='Integrity S1'), orient='h', linewidth=1, whis=1, color='#ef6d31',
           showmeans=True, 
           meanprops={"marker":"o",
                      "markerfacecolor":"white", 
                      "markeredgecolor":"black",
                      "markersize":"6"})
```

```python
#Functions to change color of text in cells that are outside of SLA
def Outside_S1(val):
    color = 'red' if val > 1 else 'black'
    return 'color: %s' % color

def Outside_S2(val):
    color = 'red' if val > 5 else 'black'
    return 'color: %s' % color

def Outside_S3(val):
    color = 'red' if val > 30 else 'black'
    return 'color: %s' % color

def Outside_S4(val):
    color = 'red' if val > 45 else 'black'
    return 'color: %s' % color
```

```python
df.filter(regex='S1').style.applymap(Outside_S1)
```

```python
df.filter(regex='S2').style.applymap(Outside_S2)
```

```python
df.filter(regex='S3').style.applymap(Outside_S3)
```

```python
df.filter(regex='S4').style.applymap(Outside_S4)
```
